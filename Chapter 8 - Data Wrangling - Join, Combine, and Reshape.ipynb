```python
# COMBINING AND MERGING DATASETS

# Database-Style DataFrame Joins

df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                   'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)

# merge or join operations combine datasets by linking rows using one or more keys
# this function uses overlapping column names as the keys
```

```python
pd.merge(df1, df2, on='key')

# or you can say it explicitly
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                   'data2': range(3)})
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')

# if both keys' name are different, then you can specify them separately

# there's some data lost which doesn't have the same key on the other key
# by default, merge uses INNER JOIN
```

```python
pd.merge(df1, df2, how='outer')

# this can specify for outer join
```

```python
# Combining Data with Overlap

a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
             index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
b = pd.Series(np.arange(len(a), dtype=np.float64),
             index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
b[-1] = np.nan
```

```python
a
```

```python
b
```

```python
np.where(pd.isnull(a), b, a)

# where(arg, true, false)
```

```python
b[:-2].combine_first(a[2:])

# combine_first will perform like the same like before
```

```python
# RESHAPING AND PIVOTING

# reshape -> rearranging tabular data

# Reshaping with Hierarchical Indexing

# Hierarchical Indexing -> consistent way to rearrange data in a DataFrame

# stack -> rotates from the columns to the rows
# unstack -> rotates from the rows into the columns

data = pd.DataFrame(np.arange(6).reshape((2,3)),
                   index=pd.Index(['Ohio', 'Colorado'], name='state'),
                   columns=pd.Index(['one', 'two', 'three'],
                                   name='number'))
```

```python
data
```

```python
result = data.stack()
```

```python
result
```

```python
result.unstack()
```

```python
result.unstack(0)

# this will rotate from column to rows
```

```python
result.unstack('state')

# this will do the same
```

```python
# Rotating 'Long' to 'Wide' Format

# long or stacked format

data = pd.read_csv('macrodata.csv')
```

```python
data
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
```

```python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
```

```python
data = data.reindex(columns=columns)
```

```python
data.index = periods.to_timestamp('D', 'end')
```

```python
ldata = data.stack().reset_index().rename(columns={0: 'value'})
```

```python
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item', 'value')
```
